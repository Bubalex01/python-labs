```python
import json
from functools import reduce
from collections import defaultdict
```

```python
# Загрузка данных
with open('countries.json', 'r') as file:
    countries = json.load(file)

with open('countries-data.json', 'r') as file:
    countries_data = json.load(file)
```

```python
# Задача 2: Сделать названия стран прописными
countries_upper = list(map(str.upper, countries))
print(countries_upper)
```

```python
# Задача 3: Отфильтровать страны, содержащие 'land'
countries_with_land = list(filter(lambda country: 'land' in country.lower(), countries))
print(countries_with_land)
```

```python
# Задача 4: Отфильтровать страны с ровно шестью символами
countries_six_letters = list(filter(lambda country: len(country) == 6, countries))
print(countries_six_letters)
```

```python
# Задача 5: Отфильтровать страны с шестью и более буквами
countries_six_or_more = list(filter(lambda country: len(country) >= 6, countries))
print(countries_six_or_more)
```

```python
# Задача 6: Отфильтровать страны, начинающиеся с буквы 'E'
countries_starting_with_e = list(filter(lambda country: country.startswith('E'), countries))
print(countries_starting_with_e)
```

```python
# Задача 7: Объединить страны в предложение
countries_selected = ['Finland', 'Sweden', 'Denmark', 'Norway', 'Iceland']
sentence = reduce(lambda acc, country: acc + (', ' if acc else '') + country, countries_selected[:-1], '') + ' и ' + countries_selected[-1] + ' являются странами Северной Европы.'
print(sentence)
```

```python
# Задача 8: Объединить map и filter для задач 2 и 3
countries_land_upper = list(map(str.upper, filter(lambda country: 'land' in country.lower(), countries)))
print(countries_land_upper)
```

```python
# Задача 9: Функция categorize_countries с использованием каррирования и замыкания
def categorize_countries(pattern):
    def filter_countries(country):
        return pattern in country.lower()
    return filter_countries

# Пример использования для стран с 'land'
countries_with_land = list(filter(categorize_countries('land'), countries))
print(countries_with_land)
```

```python
# 10.1.a Отсортировать страны по названию
sorted_by_name = sorted(countries_data, key=lambda x: x['name'])
for country in sorted_by_name:
    print(country['name'])
```

```python
# 10.1.b Отсортировать страны по столице
sorted_by_capital = sorted(countries_data, key=lambda x: x['capital'])
for country in sorted_by_capital:
    print(f"{country['name']} - {country['capital']}")
```

```python
# 10.1.c Отсортировать страны по численности населения
sorted_by_population = sorted(countries_data, key=lambda x: x['population'], reverse=True)
for country in sorted_by_population:
    print(f"{country['name']} - {country['population']}")
```

```python
# 10.2 Найти 10 наиболее распространенных языков и где их используются
amount=11
language_usage = defaultdict(list)
for country in countries_data:
    for lang in country['languages']:
        language_usage[lang].append(country['name'])

most_spoken_languages = sorted(language_usage.items(), key=lambda x: len(x[1]), reverse=True)[:amount]
print(f"Top {amount} most spoken languages:")
for lang, countries_list in most_spoken_languages:
    print(f"{lang} ({len(countries_list)} countries): {', '.join(countries_list)}\n")
```

```python
# 10.3 Найти 10 наиболее населенных стран
amount=11
top_10_populated = sorted(countries_data, key=lambda x: x['population'], reverse=True)[:amount]
print(f"Top {amount} most populated countries:")
for country in top_10_populated:
    print(f"{country['name']} - {country['population']} people")
```
