```python
#1
import random

def generate_number():
    digits = list('0123456789')
    random.shuffle(digits)
    return ''.join(digits[:4])

def get_bulls_and_cows(secret, guess):
    bulls = sum(s == g for s, g in zip(secret, guess))
    cows = sum(min(secret.count(x), guess.count(x)) for x in set(guess)) - bulls
    return bulls, cows


secret_number = generate_number()
attempts = 0

print("Добро пожаловать в игру 'Быки и Коровы'!")
print("Я загадал 4-значное число. Попытайся угадать его.")

while True:
    guess = input("Введите ваше предположение: ")
    
    if not guess.isdigit() or len(guess) != 4 or len(set(guess)) != 4:
        print("Некорректный ввод. Убедитесь, что вы ввели 4 различные цифры.")
        continue

    attempts += 1
    bulls, cows = get_bulls_and_cows(secret_number, guess)
    print(f"Быки: {bulls}, Коровы: {cows}")

    if bulls == 4:
        print(f"Поздравляем! Вы угадали число {secret_number} за {attempts} попыток.")
        break
```

```python
#3
import random


choices = ["камень", "ножницы", "бумага"]


def get_player_choice():
    while True:
        player_choice = input("Выберите: камень, ножницы или бумага: ").lower()
        if player_choice in choices:
            return player_choice
        else:
            print("Неверный выбор. Пожалуйста, выберите из: камень, ножницы, бумага.")

def get_computer_choice():
    return random.choice(choices)

def determine_winner(player, computer):
    if player == computer:
        return "Ничья!"
    elif (player == "камень" and computer == "ножницы") or \
         (player == "ножницы" and computer == "бумага") or \
         (player == "бумага" and computer == "камень"):
        return "Вы победили!"
    else:
        return "Компьютер победил!"



print("Добро пожаловать в игру 'Камень-ножницы-бумага'!")
while True:
    player_choice = get_player_choice()
    computer_choice = get_computer_choice()
    
    print(f"\nВы выбрали: {player_choice}")
    print(f"Компьютер выбрал: {computer_choice}")
    
    result = determine_winner(player_choice, computer_choice)
    print(result)
    
    play_again = input("\nХотите сыграть еще раз? (да/нет): ").lower()
    if play_again != "да":
        print("Спасибо за игру! До свидания!")
        break
```

```python
#4
def get_divisors_and_check_prime(n):
    if n <= 0 or n > 100000:
        raise ValueError("Число должно быть больше 0 и не превышать 100000")

    divisors = []
    is_prime = True
    
    for i in range(1, n + 1):
        if n % i == 0:
            divisors.append(i)
        if i > 1 and i < n and n % i == 0:
            is_prime = False

    return divisors, is_prime


try:
    number = int(input("Введите число от 1 до 100000: "))
    divisors, is_prime = get_divisors_and_check_prime(number)
    print(f"Делители числа {number}: {divisors}")
    print(f"Число {number} является простым: {is_prime}")
except ValueError as e:
    print(e)
```
